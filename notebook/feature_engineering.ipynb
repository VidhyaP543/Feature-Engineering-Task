```python
import pandas as pd
import numpy as np

df = pd.read_csv("data/raw_dataset.csv")
df.head()
```

```python
df.isnull().sum()

df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
num_cols = ['age', 'salary','experience','work_hours']

df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df['experience_salary_score'] = df['experience'] * df['salary'] 

df['avg_numeric'] = df[['age', 'salary', 'experience', 'work_hours']].mean(axis=1)

df['salary_per_hour'] = df['salary'] / df['work_hours']

df['age_group'] = pd.cut(df['age'], bins=[0,25,35,60], labels=['young','mid','senior'])
```

```python
df.to_csv("data/processed_dataset.csv", index=False)
```
